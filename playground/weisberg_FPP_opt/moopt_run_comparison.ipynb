## New derived quantities

```julia
for outputs in outputs_list

    n = length(outputs[:,1])

    outputs[:,"fLH"] = outputs[:,"Psol"] ./ outputs[:,"PLH"];
    outputs[:,"accessEC"] = 89.9 .* (outputs[:,"ne0"]./1e20).^0.5 ./ (28.0 .* outputs[:,"B0"]);
    outputs[:,"Sn"] = outputs[:,"ne0"] ./ outputs[:,"<ne>"];
    outputs[:,"ne0 / Te0^1.5"] = outputs[:,"ne0"] ./ outputs[:,"Te0"].^(3/2) / 1e17
    outputs[:,"fBS"] = outputs[:,"ip_bs"] ./ outputs[:,"ip_bs_aux_ohm"]
    outputs[:,"fNI"] = outputs[:,"ip_ni"] ./ outputs[:,"ip_bs_aux_ohm"]
    outputs[:,"C_β"] = outputs[:,"fBS"] ./ outputs[:,"βpol_MHD"] .* outputs[:,"1/ϵ"].^0.5
    outputs[:,"βn_error"] = (outputs[:,"βn"] .- outputs[:,"βn_MHD"])./outputs[:,"βn_MHD"]
    
end
```
