```julia
using Revise
using FUSE
using Plots; gr();
FUSE.logging(Logging.Info);
```

```julia
IMAS.elongation_limit(3.5)*0.8826
```

```julia
ini, act = FUSE.case_parameters(:FPP, version=:v1_demount, init_from=:scalars);

ini.equilibrium.ϵ = 1/3.5
ini.equilibrium.κ = 0.88 # set less than one so that init uses this as the f_kappa parameter
ini.equilibrium.δ = 0.7
ini.equilibrium.ζ = 0.0

ini.equilibrium.xpoints = :lower
ini.build.divertors = :lower
ini.build.symmetric = false

ini.core_profiles.T_ratio = 1/0.825

ini
```

```julia
act.ActorPFcoilsOpt.optimization_scheme = :none
act.ActorEquilibrium.symmetrize = false

act.ActorHFSsizing.do_plot = true
act.ActorPFcoilsOpt.do_plot = true
act.ActorPassiveStructures.do_plot = true

act
```

```julia
dd = IMAS.dd();
@time FUSE.init(dd, ini, act; do_plot=false);
#FUSE.digest(dd)
```

```julia
FUSE.ActorEquilibriumTransport(dd, act);
FUSE.ActorHFSsizing(dd, act);
FUSE.ActorLFSsizing(dd, act);
FUSE.ActorCXbuild(dd, act);
FUSE.ActorPFcoilsOpt(dd, act);
FUSE.ActorPassiveStructures(dd, act);
FUSE.ActorNeutronics(dd, act);
FUSE.ActorBlanket(dd, act);
FUSE.ActorDivertors(dd, act);
FUSE.ActorBalanceOfPlant(dd, act);
FUSE.ActorCosting(dd, act);
```
